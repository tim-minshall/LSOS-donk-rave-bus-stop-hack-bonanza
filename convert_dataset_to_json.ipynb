```python
import geopandas as gpd
from shapely.geometry import Point
```

```python
df = pd.read_csv("bus-stops-10-06-15.csv")
# Create GeoDataFrame with EPSG:27700
gdf = gpd.GeoDataFrame(
    df,
    geometry=gpd.points_from_xy(df['Location_Easting'], df['Location_Northing']),
    crs="EPSG:27700"
)
```

```python
post_bar = gpd.GeoSeries([Point(-0.070589, 51.588068)], crs=4326).to_crs(27700)
buffer_2km = post_bar.buffer(1000)
gdf_within_1km = gdf[gdf.geometry.within(buffer_2km.iloc[0])]
```

```python
# Convert to WGS84 for Leaflet
gdf_within_1km = gdf_within_1km.to_crs("EPSG:4326")
```

```python
# Export as GeoJSON
gdf_within_1km.to_file("bus_stops.geojson", driver="GeoJSON")
```
